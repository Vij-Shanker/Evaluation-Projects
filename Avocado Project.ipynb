### Project Description

This data was downloaded from the Hass Avocado Board website in May of 2018 & compiled into a single CSV. 
The table represents weekly 2018 retail scan data for National retail volume (units) and price. Retail scan data comes directly from retailers’ cash registers based on actual retail sales of Hass avocados. 
Starting in 2013, the table below reflects an expanded, multi-outlet retail data set. Multi-outlet reporting includes an aggregation of the following channels: grocery, mass, club, drug, dollar and military. The Average Price (of avocados) in the table reflects a per unit (per avocado) cost, even when multiple units (avocados) are sold in bags. 
The Product Lookup codes (PLU’s) in the table are only for Hass avocados. Other varieties of avocados (e.g. greenskins) are not included in this table.


Some relevant columns in the dataset:
    
•Date - The date of the observation

•AveragePrice - the average price of a single avocado

•type - conventional or organic

•year - the year

•Region - the city or region of the observation

•Total Volume - Total number of avocados sold

•4046 - Total number of avocados with PLU 4046 sold

•4225 - Total number of avocados with PLU 4225 sold

•4770 - Total number of avocados with PLU 4770 sold

#### (vi) Feature Engineering:
   - To transform variables (using Box cox transformation) to address skewness.
   - To appropriately encode categorical variables('type' and	'region').

```python
from scipy.stats import boxcox
df3=pd.DataFrame()
df2_reset=df2.reset_index(drop=True)   # reset index of df2
for i in df2_reset[Numerical]:
    coll, lambda_value = boxcox(df2_reset[i]+0.001)   # transforming each of the numerical columns of df2_reset
    df3[i]=coll
df3.skew()
```

#### Observation: 
- The skewness of all the numerical variables have considerably reduced as a result of the Box cox transformation.

#### Next
- To replace the numerical variables by their transformed values

```python
# Concatenating the numerical and categorical variables
df4 = pd.concat([df3, df2_reset[['type', 'region']]], axis=1)
df4.head()
```

#### Observation:
- All the numerical variables got replaced by their respective transformed values in the data frame.

#### Next:
- To encode categorical variable 'region' using Label Encoder. The region to use Label Encoder and not One hot encoding is that it has 54 unique values which means there would be 54 additional columns if we use one hot encoding, which will make dataframe too big.
- Apart from this "region" is the class, which is to be labelled numerically.

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
encoded=LE.fit_transform(df4['region'])
df6=df4.drop(columns='region',inplace=False)
df7=pd.concat([df6,pd.DataFrame(encoded,columns=['region'])],axis=1)
df7.tail()
```

#### Observation:
- The categorical values in "region" got replaced by integer labels.

#### Next
- Replacing the categorical variable "type" using one hot encoding. The reason to use One hot encoding is that there are only 2 unique values which will lead to addition of two additional columns. However, if we use drop_first command then only 1 additional column would be created.

```python
df8=pd.get_dummies(df7,drop_first=True)
df8.head()
```

### Model Building

```python
# Import Classification libraries
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import cross_val_score,GridSearchCV
```

- Splitting the dataset as features (X) and labels (y)

```python
X=df8[['AveragePrice', 'Total Volume', '4046', '4225', '4770', 'Total Bags',
       'Small Bags', 'Large Bags', 'XLarge Bags',
         'type_organic']]                # Features
y=df8['region']                                            # Labels
y.head()
```

```python
X2.describe()
```

#### Next:
- Random state: To find the best one

```python
from sklearn.model_selection import train_test_split
maxAcc=0
for i in range(5000):
    X_train,X_test,y_train,y_test = train_test_split(X2,y, test_size=0.3,random_state=i)
    LR=LogisticRegression(max_iter=1000)
    model=LR.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    Acc=accuracy_score(y_test,y_pred)
    if Acc>maxAcc:
        maxAcc=Acc
        random=i
print(f'Max accuracy: {maxAcc}')
print(f'Best_random_state: {random}')
```

### Logistic Regression

```python
X_train,X_test,y_train,y_test = train_test_split(X2,y, test_size=0.3,random_state=random)
LR=LogisticRegression()
model=LR.fit(X_train,y_train)
y_pred=model.predict(X_test)
Acc=accuracy_score(y_test,y_pred)
report=classification_report(y_test,y_pred)
conf_matrix=confusion_matrix(y_test,y_pred)
cv_score=cross_val_score(LR,X_train,y_train,cv=10).mean()  # performing 10 folds cross validation
print(f'Accuracy score : {round(Acc,2)}')
print(f'Cross validation score: {round(cv_score,2)}')
print(f'Classification report:\n {report}')
```

### Hyperparameter tuning of Random Forest Classifier using GridSearchCV
- To look at hyperparameters currently in use by Random Forest Classifier

```python
print('Parameters currently in use:\n')
GridSearchCV.get_params(RandomForestClassifier())
```

```python
from sklearn.model_selection import GridSearchCV
RF1=RandomForestClassifier()

hyp={'bootstrap': [True],                # each tree is built using a bootstrap sample of the training data
 'ccp_alpha': [0, 0.01, 0.1, 1.0],       # to control the complexity of the tree by pruning
 'class_weight': [None],                 # set to none as both classes are balanced
 'criterion': ['gini','entropy'],
 'max_depth': [2,6,10,14,18,20,22,24],   # maximum depth of each decision tree in the forest
 'max_features': ['auto'],
 'max_leaf_nodes': [None],
 'max_samples': [None],
 'min_impurity_decrease': [0.0],      # minimum impurity decrease required to split a node further
 'min_samples_leaf': [1],
 'min_samples_split': [2,3,5,10],
 'min_weight_fraction_leaf':[0.0],
 'n_estimators': [100,200,300],      # number of base learners (trees) to be included in the ensemble
 'n_jobs': [-1],                     # Utilize all available cores
 'oob_score': [False],               # average of the prediction accuracy on each out-of-bag sample across all trees in the forest.
 'random_state': [random],
 'verbose': [0],
 'warm_start': [True]}                # to reuse the existing solution and continue training an ensemble method from where it left off
grid_search=GridSearchCV(RF1,param_grid=hyp,cv=5)
grid_search.fit(X_train,y_train)

print("Best parameters:", grid_search.best_params_)
print("Best estimators:", grid_search.best_estimator_)
print("Best score:", round(grid_search.best_score_,3))
```

#### Observation: The hypertuned Random Forest Model has poorer accuracy (= 0.908) compared to the default "Random Forest Model" (= 0.923).
